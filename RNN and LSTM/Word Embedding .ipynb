```python
from tensorflow.keras.preprocessing.text import one_hot
```

```python
### sentences
sent=[  'the glass of milk',
     'the glass of juice',
     'the cup of tea',
    'I am a good boy',
     'I am a good developer',
     'understand the meaning of words',
     'your videos are good',]
```

```python
sent
```

```python
#this is vocubulary size
voc_size=10000
```

```python
#Used to get Index from the dictionary
one_hot_repr=[one_hot(words,voc_size)for words in sent]
print(one_hot_repr)
```

```python
from keras.layers import Embedding
from keras.models import Sequential
from tensorflow.keras.preprocessing.sequence import pad_sequences
```

```python
import numpy as np
```

```python
sent_len=8
embedding_docs=pad_sequences(one_hot_repr,padding='pre',maxlen=sent_len)
print(embedding_docs)
```

```python
model=Sequential()
model.add(Embedding(voc_size,10,input_length=sent_len))
model.compile('adam','mse')
```

```python
model.summary()
```

```python
print(model.predict(embedding_docs))
```
